```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8')
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["font.size"] = 16
plt.rc('xtick', labelsize=16)
plt.rc('ytick', labelsize=16)
plt.rc('axes', labelsize=16)
plt.rcParams["figure.autolayout"] = True

save_folders = ["saves/cleanup_baseline/episodes/episode0/", "saves/cleanup_swap_and_transfer/episodes/episode0/", "saves/game_save_20230925_001519/episodes/episode0/"]

glitch_reward_file = "Glitch_reward.txt"
gizmo_reward_file = "Gizmo_reward.txt"
glitch_rewards = []
gizmo_rewards = []

for save_folder in save_folders:
    # import from txt file with one number per line
    with open(save_folder + glitch_reward_file) as f:
        data = f.read().splitlines()
        glitch_reward = np.array(data[:120]).astype(np.float32) * 3
        glitch_rewards.append(glitch_reward)

    with open(save_folder + gizmo_reward_file) as f:
        data = f.read().splitlines()
        gizmo_reward = np.array(data[:120]).astype(np.float32) * 3
        gizmo_rewards.append(gizmo_reward)
```

```python
# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the lines
ax.plot(gizmo_rewards[0], color='tab:blue', label='Gizmo')
ax.plot(glitch_rewards[0], color='tab:green', label='Glitch')

# Filling the area between the lines
# ax.fill_between(range(len(gizmo_rewards[0])), gizmo_rewards[0], glitch_rewards[0], color='grey', alpha=0.5)

# Adding title and labels
# ax.set_title('Comparison of Rewards Over Time for Gizmo and Glitch', fontsize=16, fontweight='bold')
ax.set_xlabel('Time steps (seconds)', fontsize=14)
ax.set_ylabel('Reward (emeralds)', fontsize=14)

# # Enhancing legend
# ax.legend(frameon=True, fontsize=12)
ax.legend(frameon=True, facecolor='white', framealpha=1, edgecolor='black', fontsize=14)


# ax.legend()

# Showing the plot
plt.show()
```
